```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
## 6. Create and Plot a Line Graph from a DataFrame
## Columns:

## 'year': List of years (e.g., [2018, 2019, 2020, 2021, 2022])
## 'sales': Corresponding sales numbers (can be any integers or floats, e.g., [120, 135, 150, 145, 160])
 

df_1 = {'year': [2018, 2019, 2020, 2021, 2022], 'sales': [120, 135, 150, 145, 160]}

df_1 = pd.DataFrame(df_1, columns = ['year', 'sales'])
print(df_1)


plt.plot(df_1['year'], df_1['sales'])
plt.xticks(df_1['year']) ## this ensures only your year s are shown as x-axis labels!
plt.xlabel('Year', fontsize = 19)
plt.ylabel('Sales', fontsize = 19)
plt.title('Sales per year')
plt.show()


```

```python
## 7. Scatter Plot of DataFrame Columns
## Columns:

## 'height': Heights of at least 8 people (e.g., [150, 160, 155, 170, 165, 180, 175, 160])
## 'weight': Corresponding weights (e.g., [50, 65, 55, 72, 68, 80, 77, 62])

df_2 = {'height': [150, 160, 155, 170, 165, 180, 175, 160], 'weight': [50, 65, 55, 72, 68, 80, 77, 62]}

df_2 = pd.DataFrame(df_2, columns = ['height', 'weight'])
print(df_2)

plt.scatter(df_2['height'], df_2['weight'])
plt.xlabel('Height', fontsize = 19)
plt.ylabel('Weight', fontsize = 19)
plt.title('Weight per height')
plt.show()




```

```python
## 8. Bar Chart of Value Counts
## Columns:

## 'category': At least 12 items, with at least 4 different categories (e.g., ['A', 'B', 'C', 'D', 'A', 'B', 'C',
##  'A', 'D', 'C', 'B', 'D'])

df_3 = {'category': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'A', 'D', 'C', 'B', 'D']}
df_3 = pd.DataFrame(df_3, columns = ['category'])
print(df_3)

# Count occurences and save to a variable
counts = df_3['category'].value_counts()
print(counts)

# Plot the counts as a bar chart
counts.plot(kind = 'bar')

plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Occurences per Category')
plt.show()
```

```python
## 9. Plotting Multiple Columns on One Graph
## Columns:

## 'day': Days of the week (e.g., ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
## 'temperature': Daily temperature values (e.g., [20, 22, 19, 24, 23, 25, 21])
## 'humidity': Daily humidity percentages (e.g., [60, 65, 70, 68, 64, 63, 67])


df_4 = {'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
        'temperature': [20, 22, 19, 24, 23, 25, 21],
        'humidity': [60, 65, 70, 68, 64, 63, 67]}

df_4 = pd.DataFrame(df_4, columns = ['day', 'temperature', 'humidity'])
print(df_4)


df_4.set_index('day', inplace =True)
counts = df_4[['temperature', 'humidity']]
print(counts)

counts.plot(kind = 'bar')

plt.xlabel('Day')
plt.ylabel('Value')
plt.title('Temperature and humidity over a week')
plt.show()
```

```python
## 10. DataFrame Histogram Visualization
## Columns:

## 'ages': 20 ages, random values between 10 and 60 (e.g., [12, 15, 22, 31, 19, 25, 40, 45,
##  36, 50, 27, 33, 59, 48, 52, 20, 18, 35, 44, 60])


df_5 = {'ages': [12, 15, 22, 31, 19, 25, 40, 45, 36, 50, 27, 33, 59, 48, 52, 20, 18, 35, 44, 60]}

df_5 = pd.DataFrame(df_5, columns = ['ages'])
print(df_5)


df_5['ages'].plot(kind = 'hist', bins = 20, alpha = 0.7, edgecolor='black')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Histogram of Ages')
```
